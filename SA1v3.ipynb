```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import math as m
from typing import List

%matplotlib inline
```

```python
def f1(x_vector: List[float]) -> List[float]:
    func = np.vectorize(lambda x: (30 + 5 * x ** 2) / (24 - 2 * x))
    return func(x_vector)   
    
def f2(x_vector: List[float]) -> List[float]:
    func = np.vectorize(lambda x: 40 - 7 * x)
    return func(x_vector)
```

```python
round_ = np.vectorize(lambda x: round(x, - int(np.log10(e))))

```

```python
def interval(x1: int, x2: int, e: float) -> List[float]: 
    return list(round_(np.linspace(x1, x2, int((x2 - x1) / e) + 1)))
```

```python
x1 = 1
x2 = 6
e = 0.001
interval_ = interval(x1, x2, e)
f1res = f1(interval_)
f2res = f2(interval_)
y1 = 2.5
y2 = 22   
```

```python
plt.rcParams['figure.figsize'] = (16, 10)
plt.plot(interval_, f1res, label='line', color='green')
plt.plot(interval_, f2res, label='line', color='red')
plt.plot(interval_, [y1 for _ in range(len(interval_))], label='line', color='black')
plt.plot(interval_, [y2 for _ in range(len(interval_))], label='line', color='blue')
plt.legend(['f_1', 'f_2', 'f_1*', 'f_2*'])
```

```python
new_interval = tuple(list(np.sort([interval_[np.argmin(abs(f1res - y1))], interval_[np.argmin(abs(f2res - y2))]])) + [e])
```

```python
new_interval
```

```python
x_interval = np.arange(*new_interval)
plt.rcParams['figure.figsize'] = (8, 5)
plt.plot(x_interval, f1(x_interval), label='line', color='green')
plt.plot(x_interval, f2(x_interval), label='line', color='red')
plt.plot(x_interval, [y1 for _ in range(len(x_interval))], label='line', color='black')
plt.plot(x_interval, [y2 for _ in range(len(x_interval))], label='line', color='blue')
plt.legend(['f_1', 'f_2', 'f_1*', 'f_2*'])
```

```python
pd.set_option('display.max_rows', None)
data_frame = pd.DataFrame(x_interval, columns=['x'])
data_frame.set_index('x', inplace=True)
data_frame['f_1 / f_1*'] = f1(x_interval) / y1
data_frame['f_2 / f_2*'] = f2(x_interval) / y2
data_frame['max(f_i / f_i*)'] = data_frame[['f_1 / f_1*', 'f_2 / f_2*']].max(axis=1)
data_frame['min(max(f_i / f_i*))'] = np.where(data_frame['max(f_i / f_i*)'] == data_frame['max(f_i / f_i*)'].min(), data_frame['max(f_i / f_i*)'], '')
data_frame['min(f_i / f_i*)'] = data_frame[['f_1 / f_1*', 'f_2 / f_2*']].min(axis=1)
data_frame['max(min(f_i / f_i*))'] = np.where(data_frame['min(f_i / f_i*)'] == data_frame['min(f_i / f_i*)'].max(), data_frame['min(f_i / f_i*)'], '')
data_frame
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import math as m
from typing import List

%matplotlib inline
```

```python
def interval(x1: int, x2: int, e: float) -> List[float]: 
    return list(round_(np.linspace(x1, x2, int((x2 - x1) / e) + 1)))
round_ = np.vectorize(lambda x: round(x, - int(np.log10(e))))
```

```python
e = 0.001
num=0.01
x1=interval(-1,1,num)
x2=interval(-1,4,num)
```

```python
def f12(x1 : list, x2 : list) -> np.ndarray:
    frd=np.zeros((len(x1),len(x2)))
    for i in range(len(x1)) :
        for j in range(len(x2)) :
            frd[i][j] = 0.01 * m.sin(x1[i]) ** 2 - 0.07 * (x2[j] + (x1[i] + 4.1) * (x1[i] + 4.1))
            #frd[i][j]=-x1[i]**2+2*x1[i]+x2[j]**2-4*x2[j]+8
    return frd

def f21(x1 : list, x2 : list) -> np.ndarray :
    frd=np.zeros((len(x1), len(x2)))
    for i in range(len(x1)) :
        for j in range(len(x2)) :
            frd[i][j] = 0.01 * m.cos(x2[j])**2 + 0.3 * x1[i] * x2[j] + 0.14 * x2[j]**2+0.1 * x1[i]**2
    return frd

def f_val(x1:list, x2:list, matrix: list) -> float :
    v=np.zeros(len(x1))
    for i in range(len(x1)):
        v[i]= min(matrix[i]) #min  x2
    t=max(v) # max x1      
    return t
```

```python
f12res=f_val(x1,x2,f12(x1,x2))
f21res=f_val(x1,x2,f21(x1,x2))
```

```python
print(f12res, f21res)
```

```python
ddft=f12(x1,x2)
ddfl=f21(x1,x2)
V=[]
for i in range(len(x1)):
    for j in range(len(x2)):
        
        if (ddft[i][j] >= f12res and ddfl[i][j] >= f21res) : 
            V.append([i,j])
```

```python
for i in range(len(V)):
    plt.scatter(x1[0] + num * V[i][0], x2[0]  +num * V[i][1], color='red')
```
